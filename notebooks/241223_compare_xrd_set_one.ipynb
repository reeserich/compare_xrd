```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mineral_name = 'dolomite'
rruff_id = 'R040030'
data_df = pd.read_csv('../data/dolomite_R040030.txt', sep='\s+')
im = mpl.image.imread('../img/41598_2018_21370_Fig2_HTML.jpg')

fig = plt.figure(figsize=(10,10))
ax = plt.gca()

# calibration params
x_axis_height_pixels = 500
x_axis_start_deg= 20
x_axis_end_deg = 80
x_axis_start_pixels = 107
x_axis_end_pixels = 834

# style
multiplier = 2

def convert(convert):
    return x_axis_start_pixels + (x_axis_end_pixels - x_axis_start_pixels)*\
    (convert - x_axis_start_deg)/(x_axis_end_deg - x_axis_start_deg)

implot = ax.imshow(im, cmap='Greys_r')

# for calibration
#ax.plot([convert(20), convert(20)], [x_axis_height,x_axis_height-100], color='g', zorder=5)
#ax.plot([convert(80), convert(80)], [x_axis_height,x_axis_height-100], color='g', zorder=5)

for index, row in data_df.iterrows():
    two_theta = row['2-THETA']
    intensity = row['INTENSITY']
    if (two_theta > x_axis_start_deg) & (two_theta < x_axis_end_deg):
        ax.plot([convert(two_theta), convert(two_theta)], 
                [x_axis_height_pixels,x_axis_height_pixels-intensity*multiplier], 
                color='r', zorder=5)
        
ax.axis('off')
ax.text(s='Author labels', x=360, y=230, rotation=90, color='r', size=16)
ax.text(s='Expected ' + mineral_name + ' peak positions\n(RRUFF ID ' + rruff_id +')', 
        x=820, y=360, ha='right', rotation=0, color='r', size=16)
label_top = 255
label_bottom = label_top + 25
ax.fill_between([395, 510], 
                y1=[label_bottom,label_bottom], 
                y2=[label_top,label_top], linewidth=2, edgecolor='r', facecolor='none')
fig.savefig('../img/set_one/241223_' + mineral_name + '_' + rruff_id + 'png', dpi=200, bbox_inches='tight')
```
